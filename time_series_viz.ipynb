```python
# load Libraries
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# Load Data
df_stock_price = pd.read_csv('data/AAPL.csv', index_col=0, parse_dates=True)
```

```python
# View Data
df_stock_price.head()
```

## What is the Closing Stock Price Trend 

```python
# checking datatypeof index column 
df_stock_price.index
```

```python
# The Closing Price of Stocks over the year
plt.figure(dpi=198)
plt.plot(df_stock_price.index, df_stock_price['Close'], color='green',
          linestyle='dotted', linewidth=.7)
plt.xticks(rotation=60)
plt.xlabel('Close Date')
plt.ylabel('Close Price ($)')
plt.title('AAPL Stock Price Trend')
```

## What is the stock volume trend?

```python
plt.figure(dpi=198)
plt.bar(df_stock_price.index, df_stock_price['Volume'])
plt.xticks(rotation=90)
plt.xlabel('Date')
plt.ylabel('Volume')
plt.title('AAPL Stock Volume Trend')
```
